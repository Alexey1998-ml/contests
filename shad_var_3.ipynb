```python
res = []
def helper(curr, k, s, z):
  if len(curr) == 13:
    res.append(curr)
  if k > 0:
    if (len(curr) > 0 and curr[-1] != 'k') or len(curr) == 0:
      helper(curr + ['k'], k-1, s, z)
  if s > 0:
    helper(curr + ['s'], k, s - 1, z)
  if z > 0:
    helper(curr + ['z'], k, s, z - 1)
helper([], 5, 4, 4)
print(len(res))
```

таким образом нужно выбрать из 9 перегородок 5 мест для красных цветков (Это $ \frac{9!}{(9-5)!5!}$ ).
Нужно найти количество вариантов расположения этих звезд. это $ \frac{8!}{(8-4)!4!}$, тк цветки одного цвета считаем одинаковыми

# Task I

image.png

```python
password = input()
if len(password) < 8:
  print('no')
else:
  flag_upper = False
  flag_lower = False
  flag_digit = False
  for char in password:
    if char.islower():
      flag_lower = True
    if char.isupper():
      flag_upper = True
    if char.isdigit():
      flag_digit = True
    if flag_upper and flag_digit and flag_lower:
      break
  if flag_upper and flag_digit and flag_lower:
    print('yes')
  else:
    print('no')
```
